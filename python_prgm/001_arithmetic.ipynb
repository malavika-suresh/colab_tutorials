```
x = 100
y = 5
```

```
a = x+y       # addition of 2 numbers x and y
b = x-y       # subtract y from x
c = x*y       # to multiply x and y
d = x/y       # to divide y from x
e = x//y      # floordivison....to discard decimal part in quotient
f = x % y     # modulus.....to find remainder
g = x**y      # to find the power(exponent)
```

```
print (x,'+',y,'=',a)     # or print a
print (x,'+',y,'=',b)     # or print b
print (x,'+',y,'=',c)     # or print c
print (x,'+',y,'=',d)     # or print d
print (x,'+',y,'=',e)     # or print e
print (x,'+',y,'=',f)     # or print f
print (x,'+',y,'=',g)     # or print g  
```
